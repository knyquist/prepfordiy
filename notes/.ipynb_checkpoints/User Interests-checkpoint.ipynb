```python
import json
from pprint import pprint
```

```python
f = open('./hobby_list_wikipedia.json', 'r')
js_str = f.read()
f.close()
obj = json.loads(js_str)
pprint(obj)
```

```python
jo = { 'cats': [] }

def make_nice(x):
    has_emb = isinstance(x, list) and (True not in set([isinstance(i, dict) for i in x]))
    
    if isinstance(x, str):
        return { 'name': x }
    elif isinstance(x, list) and has_emb:
        r = []
        for i in x:
            r.append(make_nice(i))
        return r
    elif isinstance(x, list) and not has_emb:
        return { 'name': x }
    elif isinstance(x, dict):
        r = []
        for (k,v) in x.items():
            r.append({ 'name': k,
                       'subcats': [make_nice(v)] })
        return r
        
pprint(make_nice(obj))
```
